```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Steam Table.csv')
```

```python
df
```

```python
df.columns = ['Pressure','Temp','Density_f','Density_g','U_f','U_g','U_fg','H_f','H_g','H_fg','S_f','S_g','S_fg','v_f','v_g']
```

```python
df
```

```python
df.isnull().sum()
#shows how many NaN each column has
```

```python
X = df[['Pressure','Temp']].values
#y = df.drop('Pressure',axis=1).drop('Temp',axis=1).values
y = df[['U_f','U_g','H_f','H_g']].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout  #randomly turns off layers
from tensorflow.keras.callbacks import EarlyStopping
#from keras.losses import mean_squared_error
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min',verbose=1, patience=20)
```

```python
model = Sequential()

model.add(Dense(500,activation='relu'))
model.add(Dense(500,activation='relu'))
model.add(Dense(500,activation='relu'))

model.add(Dense(4))
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,epochs=10000,batch_size=16,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
dfloss = pd.DataFrame(model.history.history)
```

```python
dfloss.plot()
```

```python
np.sqrt(model.evaluate(X_train,y_train))
```

```python
np.sqrt(model.evaluate(X_test,y_test))
```

```python
pred = model.predict(X_test)
```

```python
predictions = pd.DataFrame(pred)
predictions.columns = ['Pred_U_f','Pred_U_g','Pred_H_f','Pred_H_g']
predictions.head()
```
